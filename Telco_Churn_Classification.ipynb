```python
import pandas as pd #importing pandas 

data = pd.read_csv('C:/Users/user/Documents/WA_Fn-UseC_-Telco-Customer-Churn.csv') #I downloaded this csv file from Kaggle
data
```

```python
data.pop('customerID') #Deleting customerID
data
```

```python
from sklearn.preprocessing import LabelEncoder #importing LabelEncoder in order for us to change the encoding of values into integers

#data.columns cause we are going to drop all the values after changing it to integers
drop = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod',
       'MonthlyCharges', 'TotalCharges', 'Churn']

#Creating an instance of LabelEncoder
gender = LabelEncoder()
Partner = LabelEncoder()
Dependents = LabelEncoder()
tenure = LabelEncoder()
MultipleLines = LabelEncoder()
InternetService = LabelEncoder()
OnlineSecurity = LabelEncoder()
OnlineBackup = LabelEncoder()
DeviceProtection = LabelEncoder()
TechSupport = LabelEncoder()
StreamingTV = LabelEncoder()
StreamingMovies = LabelEncoder()
Contract = LabelEncoder()
PaperlessBilling = LabelEncoder()
PaymentMethod = LabelEncoder()
PhoneService = LabelEncoder()
MonthlyCharges = LabelEncoder()
TotalCharges = LabelEncoder()
Churn = LabelEncoder()

#We are now adding a new sets of columns with integers as a values
data['GENDER'] = gender.fit_transform(data['gender'])
data['PARTNER'] = gender.fit_transform(data['Partner'])
data['DEPENDENTS'] = gender.fit_transform(data['Dependents'])
data['TENURE'] = gender.fit_transform(data['tenure'])
data['MULTIPLELINES'] = gender.fit_transform(data['MultipleLines'])
data['INTERNETSERVICE'] = gender.fit_transform(data['InternetService'])
data['ONLINESECURITY'] = gender.fit_transform(data['OnlineSecurity'])
data['ONLINEBACKUP'] = gender.fit_transform(data['OnlineBackup'])
data['DEVICEPROTECTION'] = gender.fit_transform(data['DeviceProtection'])
data['TECHSUPPORT'] = gender.fit_transform(data['TechSupport'])
data['STREAMINGTV'] = gender.fit_transform(data['StreamingTV'])
data['STREAMINGMOVIES'] = gender.fit_transform(data['StreamingMovies'])
data['CONTRACT'] = gender.fit_transform(data['Contract'])
data['PAPERLESSBILLING'] = gender.fit_transform(data['PaperlessBilling'])
data['PAYMENTMETHOD'] = gender.fit_transform(data['PaymentMethod'])
data['PHONESERVICE'] = gender.fit_transform(data['PhoneService'])
data['MONTHLYCHARGES'] = gender.fit_transform(data['MonthlyCharges'])
data['TOTALCHARGES'] = gender.fit_transform(data['TotalCharges'])
data['CHURN'] = gender.fit_transform(data['Churn'])

#droping the old columns 
cleaned_data = data.drop(columns=drop)

#finally we now have a cleaned data
cleaned_data
```

```python
#splitting the cleaned data into training data and test_data 
training_data = cleaned_data[1:6044] #from 1 - 6043 
test_data = cleaned_data[6044:7044] #from 6044 - 7043 or the last 1000 data

#the training input and training target
training_input = training_data.drop(columns="CHURN")
training_target = training_data['CHURN']

#the test input and test target
test_input = test_data.drop(columns="CHURN")
test_target = test_data['CHURN']
```

###### DecisionTreeClassifier

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(training_input, training_target)
score = model.score(test_input, test_target)
print("Accuracy Score of DecisionTreeClassifier is {} %".format(score * 100))
```

###### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(4)
model.fit(training_input, training_target)
score = model.score(test_input, test_target)
print("Accuracy Score of KNeighborsClassifier is {} %".format(score * 100))
```

###### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(10)
model.fit(training_input, training_target)
score = model.score(test_input, test_target)
print("Accuracy Score of RandomForestClassifier is {} %".format(score * 100))
```

###### ExtraTreeClassifier

```python
from sklearn.tree import ExtraTreeClassifier

model = ExtraTreeClassifier()
model.fit(training_input, training_target)
score = model.score(test_input, test_target)
print("Accuracy Score of ExtraTreeClassifier is {} %".format(score * 100))
```

###### LinearDiscriminantAnalysis

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

model = LinearDiscriminantAnalysis()
model.fit(training_input, training_target)
score = model.score(test_input, test_target)
print("Accuracy Score of LinearDiscriminantAnalysis is {} %".format(score * 100))
```

###### GaussianNB

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(training_input, training_target)
score = model.score(test_input, test_target)
print("Accuracy Score of GaussianNB is {} %".format(score * 100))
```

###### BaggingClassifier

```python
from sklearn.ensemble import BaggingClassifier

model = BaggingClassifier()
model.fit(training_input, training_target)
score = model.score(test_input, test_target)
print("Accuracy Score of BaggingClassifier is {} %".format(score * 100))
```
